# Testing stability between elliptical and trapezoidal fin shapes

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['figure.titlesize'] = 20
plt.rcParams['figure.titleweight'] = 'bold'
plt.rcParams['figure.figsize'] = (10, 8)
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['figure.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 12
plt.rcParams['mathtext.fontset'] = 'cm'
```

```python
trapezoidal = pd.read_csv('Trapezoidal Stability.csv', names=['Time (s)', 'Stability Margin'])
elliptical = pd.read_csv('Elliptical Stability.csv', names=['Time (s)', 'Stability Margin'])

trapezoidal_times = trapezoidal['Time (s)']
elliptical_times = elliptical['Time (s)']

trapezoidal_stability = trapezoidal['Stability Margin']
elliptical_stability = elliptical['Stability Margin']
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(trapezoidal_times, trapezoidal_stability, label='Trapezoidal Fins', color='blue')
ax.plot(elliptical_times, elliptical_stability, label='Elliptical Fins', color='orange')

ax.vlines(x=0.29, ymin=0, ymax=10, colors='green', linestyles='dashed', label='Launch Rod Exit')
ax.vlines(x=2.309, ymin=0, ymax=10, colors='red', linestyles='dashed', label='Motor Burnout')
ax.vlines(x=17.351, ymin=0, ymax=10, colors='purple', linestyles='dashed', label='Apogee')

ax.set_title('Stability Margin Comparison')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Stability Margin')

ax.set_ylim(1.75, 3.75)

ax.legend(loc='lower center')
ax.grid(True)

plt.show()
```
